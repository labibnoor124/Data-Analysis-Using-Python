```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot 1: Simple numerical x-y plot
x = np.array([1, 2, 3, 4])
y = np.array([10, 20, 15, 25])
plt.plot(x, y)
plt.show()
```

Data Point -----> Each individual pairs in our data set is data point.

X-axis ----> Horizontal,input data.

Y-axis ----> Vertical,output data.

Plot ------> the line,shape that is actually our graph.

Marker -----> a symbol or dot that is used to make each data point visible.

o-circle,s-square,d-diamond,x-cross

Line Style ----> connection between each data point.

"-" --> solid line,"--" --> dashed line,":"-->dottet line,"-." ---> dashed dot lines

Legend ----> dedicated box to tell us which line means what.

Label -----> tells us what is the purpose of each axis.


```python
# Plot 2: Days of the week vs sales
x = np.array(["Mon", "Tue", "Wed", "Thu", "Fri"])
y = np.array([10, 15, 7, 20, 12])
plt.plot(x, y)
plt.title("Sales Analysis")
plt.xlabel("Day of the week")
plt.ylabel("Sales Per Day")
plt.show()
```

```python
# Plot 3: Monthly sales with styling
months = np.array(["Jan", "Feb", "Mar", "Apr", "May", "Jun", 
                   "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])
sales = np.array([12000, 14500, 8900, 7500, 16000, 17800, 
                  21000, 14560, 16700, 7900, 6700, 12000])
plt.plot(months, sales, color='red', linestyle="--", linewidth=3, marker="o",label="2023 sales Data")
plt.xlabel("Months")
plt.ylabel("Sales Per Month")
plt.title("Monthly Sales of a Shop")
plt.legend(loc="upper right",fontsize=10)
plt.grid(color='blue',linestyle=':',linewidth=2)
plt.show()
```

Scatter plots are used in inconsistent data where we need to find a trend

plt.scatter(x,y,color,marker,lebel)

```python
hours_studies=np.array([3,2.5,1,7,5,6.4,9,8.6])
exam_scores=np.array([53,59,32,56,67,71,77,61])
plt.scatter(hours_studies,exam_scores,color="red",marker="o",label="Marks Analysis")
plt.xlabel("Hours Studied")
plt.ylabel("Marks Scored")
plt.legend()
plt.title("Marks Analysis Against Hours Studied")
plt.grid(True)
plt.show()
```

```python
hours=np.array([1,2,3,4])
class1=np.array([34,45,51,58])
class2=np.array([37,42,46,61])
plt.scatter(hours,class1,color="red",label="Class 1")
plt.scatter(hours,class2,color="blue",label="Class 2")
plt.legend()
plt.xlabel("Hours Studied")
plt.ylabel("Marks Obtained")
plt.grid(True)
plt.title("Comparison Between 2 classes")
plt.show()
```

Subplot:

#easy way

plt.subplot(nrows,ncols,index)

here the index will be one based not zero based

```python
x=np.array([1,2,3,4])
y=np.array([7,12,9,11])


plt.subplot(1,2,1)
plt.plot(x,y,color="red",linestyle="--",marker="o",linewidth=3)
plt.xlabel("Hours")
plt.ylabel("Output")
plt.title("Line Chart")


plt.subplot(1,2,2)
plt.bar(x,y,color="blue")
plt.title("Bar Chart")
plt.xlabel("Hours")
plt.ylabel("Output")

plt.show()
```

#professional way

fig,ax=plt.subplots(nrows,ncols,figure_size=(width,height))



```python
fig,ax=plt.subplots(figsize=(10,5))
x=np.array([1,2,3,4])
y=np.array([10,20,15,22])
ax.plot(x,y,color="red",linestyle="--",marker="o",linewidth=2)
ax.grid(color="black",linestyle=":")
ax.set_title("Line Plot")
plt.show()
```

```python
fig,ax=plt.subplots(2,2,figsize=(12,10))
x=np.array([1,2,3,4])
y=np.array([10,20,15,22])


ax[0,0].plot(x,y,color="red",linewidth=2,linestyle="--",marker="o",label="Comparison")
ax[0,0].grid(color="black",linestyle=":")
ax[0,0].set_title("Line Chart")
ax[0,0].legend(loc="upper left",fontsize=8)
ax[0,0].set_xlabel("Input")
ax[0,0].set_ylabel("Output")

ax[0,1].bar(x,y,color="orange",label="Comparison")
ax[0,1].grid(color="black",linestyle=":")
ax[0,1].legend(loc="upper left",fontsize=8)
ax[0,1].set_title("Bar Chart")
ax[0,1].set_xlabel("Input")
ax[0,1].set_ylabel("Output")


ax[1,0].pie(y,labels=x,colors=["gold","skyblue","lightgreen","coral"],autopct="%1.1f%%")
ax[1,0].grid(color="black",linestyle=":")
ax[1,0].legend(loc="upper left",fontsize=8)
ax[1,0].set_title("Pie Chart")



plt.tight_layout()
plt.show()
```
