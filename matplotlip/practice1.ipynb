```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4])
y=np.array([10,20,15,30])
plt.plot(x,y)
plt.show()
```

```python
x=np.array(["Mon","Tue","Wed","Thu","Fri"])
y=np.array([10,15,7,20,12])
# plt.plot(x,y)
plt.plot(x,y,color="red",linestyle="-",linewidth=2,marker="o",label="2024 Sales Data")
plt.grid(color="blue",linestyle=":")
plt.title("Sales Analysis")
plt.xlabel("Days of the Week")
plt.ylabel("Sales Per Day")
plt.legend(loc="upper left",fontsize=10)
plt.show()
```

```python
x=np.array(["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"])
y=np.array([21.67,23.45,29.81,31.45,37.89,35.59,36.43,38.86,32.45,27.91,25.88,23.05])

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
plt.plot(x,y,marker="o",color="red",linestyle="-",linewidth=1,label="Average Temperature")
plt.title("Average Temperature of Year 2024")
plt.legend(loc="upper left",fontsize=8)
plt.grid(True,color='black',linestyle=":")
plt.show()

plt.figure(figsize=(12,6))
plt.subplot(1,2,2)
plt.bar(x,y,color="red")

plt.tight_layout()
plt.show()
```

```python
fig,ax=plt.subplots(1,2,figsize=(12,6))

x=np.array([1,2,3,4])
y=np.array([10,20,15,25])

ax[0].plot(x,y,color="green",linestyle="-",linewidth=2,marker="o",label="Line Chart")
ax[0].set_title("Line Chart")
ax[0].grid(True,color="black",linestyle=":")
ax[0].set_xlabel("Hours")
ax[0].set_ylabel("Marks")
ax[0].legend(loc="upper left",fontsize=8)

ax[1].bar(x,y,color="orange",label="Bar Chart")
ax[1].set_title("Bar Chart")
ax[1].grid(True,color="black",linestyle=":")
ax[1].set_xlabel("Hours")
ax[1].set_ylabel("Marks")
ax[1].legend(loc="upper left",fontsize=8)



plt.suptitle("Comparison Between Line Chart and Bar Chart")
plt.tight_layout();
plt.show()
```

To save graphs from figures we use savefig() function

plt.savefig(filename with format,dpi=value,bbox_inches="tight")



```python
x=np.array([1,2,3,4])
y=np.array([10,20,15,25])

plt.plot(x,y,color="green",marker="o",linestyle="-",linewidth=2,label="Bar Chart")
plt.xlabel("Hours")
plt.ylabel("Output")
plt.title("Simple Line Chart")
plt.grid(True,color="black",linestyle=":")
plt.legend(loc="upper left",fontsize=8)

plt.tight_layout()
plt.savefig("Simple_Line_Chart.png",dpi=300,bbox_inches="tight")
plt.show()

```
